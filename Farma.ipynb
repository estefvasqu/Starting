```
# Indexar desde 1 
pred1 = df.loc[:50, ['Costo', 'Sector']]
pred1.index += 1
pred1.head(60)
```

```
pred1.describe()
```

Manejo de los Datos con condicionales 


```
# np.where(condition, value if condition is true, value if condition is false)
pred1['Caro'] = np.where(pred1['Costo'] > 200, True, False)
pred1.head(20)
```

```
# df.loc[df['column name'] condition, 'new column name'] = 'value if condition is met'
pred1.loc[pred1['Caro'] == True, 'T or F'] = 1   
pred1.loc[pred1['Caro'] == False, 'T or F'] = 0   
pred1.head()
```

```
# Condicion if con or 
pred1.loc[(pred1['T or F'] == 0) | (pred1['T or F'] == 1), 'Obs'] = 'farma'  
pred1.head()
```

```
# Borrando una columna en un dataframe 
del pred1['Obs']
pred1.head()
```

```
pred1.loc[pred1['T or F'] == 1, 'T or F'] = 2
pred1.head()
```

```
# Bucle/loop for in 
for ind in pred1.index: 
     print(pred1['Costo'][ind], pred1['Sector'][ind]) 
```

```
# Visualizacion 
pred1.plot()
plt.show()
```
